```python
#Question 1a
def findvowels(name):
    vow = "aeiou"
    count =0
    for char in name:
        if vow.find(char.lower()) >= 0 :
            count += 1
    return count

findvowels("Hello")
```

```python
#Question 2
def count_num(name):
    my_dict = {'A':1,'B':2,'C':3,'D':4, 'E':5,'F':6,'G':7,'H':8,'I':9,'J':10,'K':11,'L':12,'M':13
          ,'N':14,'O':15,'P':16,'Q':17,'R':18,'S':19,'T':20,'U':21,'V':22,'W':23,'X':24,'Y':25,'Z':26}
    sum_1 = 0
    name = name.upper()
    for char in name:
        sum_1 = sum_1 + my_dict.get(char,0)
    return sum_1

count_num("Regards")
```

```python
#Question 3d
df_after_drop = df_Bank.drop(['job', 'day', 'month', 'pdays', 'previous','poutcome'], axis=1, inplace=False)
df_after_drop.head()
```

```python
#Question 3e, recode variables with 2 levels as 0 and 1
df_recode = df_after_drop.copy()

housing_new = np.zeros(df_recode.shape[0])

for i in range(df_recode.shape[0]):
    if df_recode.housing[i]=='yes': 
        housing_new[i]=1
    else:
        housing_new[i]=0
        
df_recode['housing_new'] = housing_new

loan_new = np.zeros(df_recode.shape[0])

for i in range(df_recode.shape[0]):
    if df_recode.loan[i]=='yes': 
        loan_new[i]=1
    else:
        loan_new[i]=0

df_recode['loan_new'] = loan_new

y_new = np.zeros(df_recode.shape[0])
for i in range(df_recode.shape[0]):
    if df_recode.y[i]=='yes': 
        y_new[i]=1
    else:
        y_new[i]=0

df_recode['y_new'] = y_new

default_new = np.zeros(df_recode.shape[0])
for i in range(df_recode.shape[0]):
    if df_recode.default[i]=='yes': 
        default_new[i]=1
    else:
        default_new[i]=0

df_recode['default_new'] = default_new
```

```python
df_recode.head()
```

```python
#Question 3f, dummification of categorical variables having more than one values
Education = pd.get_dummies(df_recode.education)
Education.columns = ["Edu1","Edu2","Edu3","Edu4"]
df_recode = df_recode.join(Education)
Contact = pd.get_dummies(df_recode.contact)
Contact.columns = ["Con1","Con2","Con3"]
df_recode = df_recode.join(Contact)
Marital = pd.get_dummies(df_recode.marital)
Marital.columns = ["Mar1","Mar2","Mar3"]
df_recode = df_recode.join(Marital)
```
